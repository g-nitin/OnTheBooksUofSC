```python
from nltk.tokenize import sent_tokenize
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('1881_Acts.txt', encoding='utf8') as f:
    acts = sent_tokenize(f.read())
```

```python
df = pd.DataFrame(acts)
df
```

```python
df = df.rename(columns= {0:'sentence'})
```

```python
df.head()
```

```python
for i in df['sentence']:
    if len(i) <=13:
        print(i)
```
